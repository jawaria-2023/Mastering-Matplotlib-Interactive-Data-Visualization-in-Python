```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

# Create a figure and an axes.
fig, ax = plt.subplots()

# Plotting the line
ax.plot(x, y)

# Adding title and labels
ax.set_title('Simple Line Plot')
ax.set_xlabel('x')
ax.set_ylabel('y = x^2')

# Show the plot
plt.show()
```

```python
# Data
platforms = ['Facebook', 'Instagram', 'Twitter', 'LinkedIn', 'Snapchat']
users = [274, 121, 396, 310, 238]  # in millions

# Creating the bar chart
fig, ax = plt.subplots()
ax.bar(platforms, users)

# Adding title and labels
ax.set_title('Social Media Users in 2023')
ax.set_xlabel('Platform')
ax.set_ylabel('Users (millions)')

# Show the plot
plt.show()
```

```python
import numpy as np

# Generating random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
area = (30 * np.random.rand(50))**2  # Bubble sizes

# Scatter plot
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, s=area, c=colors, alpha=0.5)

# Adding title and labels
ax.set_title('Random Scatter Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```
